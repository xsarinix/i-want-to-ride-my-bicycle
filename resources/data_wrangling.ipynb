```python
import pandas as pd
import os
```

```python
csv_list = os.listdir('/users/sarah/onedrive/documents/boot camp/7.tableau/homework/resources/trip_data/2018')
df_list = []
for csv in csv_list:
    filepath = 'trip_data/2018/' + csv
    dataframe = pd.read_csv(filepath)
    df_list.append(dataframe)
```

```python
frame = pd.concat(df_list, axis=0, ignore_index=True)
```

```python
frame.to_csv('/users/sarah/onedrive/documents/boot camp/7.tableau/homework/resources/trip_data/2018/2018-citibike-tripdata.csv')
```

```python
frame.head()
```

```python
starts = frame[['start station name', 'start station latitude', 'start station longitude']].rename(index=str, columns={'start station name':'station name', 'start station latitude':'station latitude', 'start station longitude':'station longitude'})
ends = frame[['end station name', 'end station latitude', 'end station longitude']].rename(index=str, columns={'end station name':'station name', 'end station latitude':'station latitude', 'end station longitude':'station longitude'})
locations = [starts, ends]
```

```python
locations_groupby=pd.concat(locations, sort=False).groupby('station name')
locations_coordinates = locations_groupby.mean()
locations_uses = locations_groupby.count().rename(index=str, columns={'station latitude':'use_count'})
locations_coordinates['use_count'] = locations_uses['use_count']
#locations_coordinates.to_csv('/users/sarah/onedrive/documents/boot camp/7.tableau/homework/resources/locations.csv', index=True)
```

```python
starts_groupby=starts.groupby('station name')
starts_coordinates = starts_groupby.mean()
starts_uses = starts_groupby.count().rename(index=str, columns={'station latitude':'use_count'})
starts_coordinates['use_count'] = starts_uses['use_count']
starts_coordinates.to_csv('/users/sarah/onedrive/documents/boot camp/7.tableau/homework/resources/starts.csv')
starts_coordinates.head()
```

```python
ends_groupby=ends.groupby('station name')
ends_coordinates = ends_groupby.mean()
ends_uses = ends_groupby.count().rename(index=str, columns={'station latitude':'use_count'})
ends_coordinates['use_count'] = ends_uses['use_count']
ends_coordinates.to_csv('/users/sarah/onedrive/documents/boot camp/7.tableau/homework/resources/ends.csv')
ends_coordinates.head()
```
